```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
```

```python
df = pd.read_csv('spam.csv', encoding='latin-1')
df.head()
```

```python
df = df[['v1', 'v2']]
```

```python
df.head()
```

```python
df.rename(columns={'v1': 'Spam', 'v2': 'Text'},inplace =True)
```

```python
df.head()
```

```python
tfid = TfidfVectorizer()
x = tfid.fit_transform(df['Text'])
```

```python
df1 = pd.DataFrame(x.toarray(), columns=tfid.get_feature_names())
```

```python
df1.head()
```

```python
nmf = NMF(n_components = 25)
nmf.fit(x)
x = nmf.transform(x)
```

```python
final_df = pd.DataFrame(x)
final_df.head()
```

```python
final_df['Spam'] = df['Spam']
final_df.head()
```

```python
final = pd.get_dummies(final_df, drop_first=True, columns=['Spam'])
final.head()
```

```python
final.rename(columns = {'Spam_spam': 'Spam'}, inplace = True)
final.head()
```

```python
final['Spam'].value_counts()
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
final.iloc[:,0:25]
```

```python
X_train, X_test, y_train, y_test = train_test_split(final.iloc[:,0:30].values, final['Spam'].values)
```

```python
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
X_train.shape
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
model.score(X_train, y_train)
```
